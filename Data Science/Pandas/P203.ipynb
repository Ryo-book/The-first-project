```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.available

df['TIME_str'] = df['<TIME>'].astype(str).str.zfill(4)  # 0 -> 0000, 5 -> 0005, 830 -> 0830


df['datetime'] = pd.to_datetime(df['<DTYYYYMMDD>'].astype(str) + df['TIME_str'], format='%Y%m%d%H%M')
df = df.set_index('datetime')
df = df.rename(columns={"<CLOSE>": "close"})


# 終値をプロット
plt.figure(figsize=(12,5))
df['close'].plot()
plt.title("USDJPY Close Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show();
```

```python
print(df.columns.tolist())
```

```python
# 列名前後の空白を削除
df.columns = df.columns.str.strip()

# <TIME> 列を文字列にして4桁に揃える
df['TIME_str'] = df['<TIME>'].astype(str).str.zfill(4)

# 日付と時間を結合して datetime に変換
df['datetime'] = pd.to_datetime(df['<DTYYYYMMDD>'].astype(str) + df['TIME_str'],
                                format='%Y%m%d%H%M')

# datetime をインデックスに
df = df.set_index('datetime')

# 終値列を 'close' に変更
df = df.rename(columns={"<CLOSE>": "close"})
df['close'] = pd.to_numeric(df['close'])  # 念のため数値型に変換
```

```python
plt.figure(figsize=(12,5))

# 元のデータ（M5ベース）
df['close'].plot(alpha=0.5, style='-')

# 年末平均でリサンプル（resample）したデータ
df['close'].resample('BA').mean().plot(style=':')

# 年末の値を取得する（asfreq）データ
df['close'].asfreq('BA').plot(style='--')

# 凡例
plt.legend(['input', 'resample', 'asfreq'], loc='upper left')

plt.title("USDJPY Close Price Comparison")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
df = df.asfreq('D', method='pad')

ROI = 100 * (df.shift(-365) - df) / df
ROI.plot()

plt.ylabel('% Return on Investment after 1 year');
```
