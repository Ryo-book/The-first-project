```python
USDJPY.head()
```

```python
USDJPY["HIGH_bin"] = pd.cut(USDJPY["<HIGH>"], bins=50)
USDJPY["LOW_bin"] = pd.cut(USDJPY["<LOW>"], bins=50)

party_counts = pd.crosstab(USDJPY["HIGH_bin"], USDJPY["LOW_bin"])
```

```python
party_counts = party_counts.reindex(index=["Thur", "Fri", "Sat", "Sun"])
party_counts
```

```python
party_pcts = party_counts.div(party_counts.sum(axis="columns"),
                              axis="index")
```

```python
party_pcts
```

```python
party_pcts.plot.bar(stacked=True)
```

```python
USDJPY = pd.read_csv(r"C:\Users\Ryouh\OneDrive\\VScode\Introduction matplotlib\USDJPY_M5.csv")
```

```python
data = USDJPY[["<TIME>", "<OPEN>", "<HIGH>", "<LOW>"]]
```

```python
trans_data = np.log(data).diff().dropna()
```

```python
trans_data.tail()
```

```python
import seaborn as sns
```

```python
ax = sns.regplot(x="<OPEN>", y="<LOW>", data=trans_data)
ax.set_title("Changes in log(OPEN) versus log(LOW)")  # 
plt.show()
```

```python
sns.pairplot(trans_data, diag_kind="kde", plot_kws={"alpha": 0.2})
```
